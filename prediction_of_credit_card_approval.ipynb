```python
import pandas as pd
train_data=pd.read_csv('train_data.csv')
print(train_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('train_data.csv')

# Display summary statistics
print("Summary Statistics:")
print(data.describe())

# Univariate Analysis
# Plot histograms for numerical features
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Plot bar plots for categorical features
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(data[feature])
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Bivariate Analysis
# Plot pair plots for numerical features
plt.figure(figsize=(10, 8))
sns.pairplot(data[numerical_features])
plt.suptitle('Pair Plot of Numerical Features')
plt.show()

# Plot heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Business Insights
```

```python
# Fill missing values with the mode (most frequently occuring data)
mode_value = train_data['Job title'].mode()[0]
train_data['Job title'].fillna(mode_value, inplace=True)
```

```python
null_values = train_data.isnull().sum()
print(null_values)
```

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le=LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in train_data.columns.to_numpy():
    # Compare if the dtype is object
    if train_data[col].dtypes=='object':
    # Use LabelEncoder to do the numeric transformation
        train_data[col]=le.fit_transform(train_data[col])
```

```python
import pandas as pd
test_data=pd.read_csv('test_data.csv')
print(test_data.head())
```

```python
# Fill missing values with the mode
mode_value1 = test_data['Job title'].mode()[0]
test_data['Job title'].fillna(mode_value, inplace=True)
```

```python
null_values1 = test_data.isnull().sum()
print(null_values1)
```

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Instantiate LabelEncoder
le=LabelEncoder()

# Iterate over all the values of each column and extract their dtypes
for col in test_data.columns.to_numpy():
    # Compare if the dtype is object
    if test_data[col].dtypes=='object':
    # Use LabelEncoder to do the numeric transformation
        test_data[col]=le.fit_transform(test_data[col])
```

```python
X_train1 = train_data.iloc[:, :-1]  # Features (all columns except the last one)
y_train1 = train_data.iloc[:, -1]  #Target[last column]
X_test1 = test_data.iloc[:, :-1]  # Features (all columns except the last one)
y_test1 = test_data.iloc[:, -1]   # Target (last column)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score

# Assuming X_train, y_train, X_test, y_test are your feature and target variables
# obtained after preprocessing your train and test datasets

# Initialize and train the model on the training dataset
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train1, y_train1)

# Make predictions on the test dataset
y_pred = model.predict(X_test1)

# Evaluate the model
accuracy = accuracy_score(y_test1, y_pred)
precision = precision_score(y_test1, y_pred)
recall = recall_score(y_test1, y_pred)
f1 = f1_score(y_test1, y_pred)
roc_auc = roc_auc_score(y_test1, y_pred)
conf_matrix = confusion_matrix(y_test1, y_pred)
print("Accuracy:", accuracy)
print("Precision:",precision)
print("Recall:", recall)
print("F1-score:", f1)
print("ROC AUC Score:", roc_auc)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model on the training dataset
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train1, y_train1)

# Make predictions on the test dataset
y_pred_logistic = logistic_model.predict(X_test1)

# Evaluate the Logistic Regression model
accuracy_logistic = accuracy_score(y_test1, y_pred_logistic)
precision_logistic = precision_score(y_test1, y_pred_logistic)
recall_logistic = recall_score(y_test1, y_pred_logistic)
f1_logistic = f1_score(y_test1, y_pred_logistic)
roc_auc_logistic = roc_auc_score(y_test1, y_pred_logistic)
conf_matrix_logistic = confusion_matrix(y_test1, y_pred_logistic)

# Print evaluation metrics
print("Evaluation metrics for Logistic Regression model:")
print("Accuracy:", accuracy_logistic)
print("Precision:", precision_logistic)
print("Recall:", recall_logistic)
print("F1-score:", f1_logistic)
print("ROC AUC Score:", roc_auc_logistic)
print("Confusion Matrix:")
print(conf_matrix_logistic)
```
